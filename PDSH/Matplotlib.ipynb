```python
# plot

import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

fig, sub = plt.subplots(2,1);
x = np.linspace(0,3*np.pi,30);
sub[0].plot(x,np.sin(x),'ok',label='sin(x)');
sub[1].plot(x,np.cos(x),'-.',label='cos(x)');
fig.legend(loc='center right');
sub[0].set(xlim=(0, 3*np.pi),ylim=(-2, 2),xlabel='x',ylabel='sin(x)',title='A Simple Plot');
sub[1].set(xlim=(0, 3*np.pi),ylim=(-2, 2),xlabel='x',ylabel='cos(x)',title='A Simple Plot 2');
fig.subplots_adjust(hspace=1.0);
```

```python
# scatter

rng = np.random.RandomState(seed=0)
n = 50
x = rng.rand(n)
y = rng.rand(n)
clr = rng.rand(n)
sz = 1000*rng.rand(n)

plt.scatter(x,y,c=clr,s=sz,cmap='viridis',alpha=0.3);
plt.colorbar();
plt.axis('equal');
plt.grid('on');
```

```python
# contour

def f(x,y):
    return np.sin(x)**2+np.cos(x*y+10)*np.cos(x)

x = np.linspace(0,4,40)
y = np.linspace(-2,1,40)
X, Y = np.meshgrid(x,y)
Z = f(X,Y)

cntrs = plt.contour(X,Y,Z,3,colors='black');
plt.clabel(cntrs, inline=True, fontsize=8);
plt.imshow(Z, extent=[0, 4, -2, 1], origin='lower',cmap='RdGy', alpha=0.5)
plt.colorbar();
```
