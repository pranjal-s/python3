```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('http://bit.ly/autompg-csv')
hp = np.array(data['hp'])
```

```python
print('Mean: %s' % hp.mean())
```

```python
sns.set()
plt.hist(hp)
plt.title('Horse Power of Cars');
plt.xlabel('horse power');
plt.ylabel('number');
```
